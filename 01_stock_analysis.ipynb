```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tickers=['AAPL','TSLA','NVDA']
```

```python
data=yf.download(
    tickers,
    start='2020-01-01',
    auto_adjust=True
)['Close']
```

```python
print(data.head())
```

```python
data.plot(title="Stock Price Comparison ")
plt.show()
```

```python
returns=data.pct_change(fill_method=None)
```

```python
returns.plot(subplots=True,
             figsize=(12,8),
             title='Daily Returns')
plt.tight_layout()
plt.show()
```

```python
total_return=data.iloc[-2]/data.iloc[0]-1
volatility=returns.std()
```

```python
print('Total_return')
print(total_return)
```

```python
print('Volatility')
print(volatility)
```

```python
aapl=data['AAPL']
ma20=aapl.rolling(20).mean()
ma60=aapl.rolling(60).mean()
plt.figure(
           figsize=(14,7))
plt.plot(aapl,label="AAPL",linewidth=1.2)
plt.plot(ma20,label="MA20",linewidth=2.2)
plt.plot(ma60,label="MA60",linewidth=2.2)

plt.title('AAPL Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
nvda=data['NVDA']
ma20=nvda.rolling(20).mean()
ma60=nvda.rolling(60).mean()
plt.figure(
           figsize=(14,7))
plt.plot(aapl,label="NVDA",linewidth=1.2)
plt.plot(ma20,label="MA20",linewidth=2.2)
plt.plot(ma60,label="MA60",linewidth=2.2)

plt.title('NVDA Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
