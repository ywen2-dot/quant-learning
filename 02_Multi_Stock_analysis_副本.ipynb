```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tickers=['AAPL','TSLA','NVDA','MSFT','GOOGL']
```

```python
data=yf.download(
    tickers,
    start='2020-01-01',
    auto_adjust=True,
    progress=False
)['Close']
```

```python
data.head()
```

```python
normalized=data/data.iloc[0]
```

```python
normalized.plot(figsize=(12,6),title='Normalized Stocke Performance')
plt.ylabel('Growth pf $1')
```

```python
returns=data.pct_change()
total_returns=data.iloc[-1]/data.iloc[0]-1
volatility=returns.std()
```

```python
print(returns)
print(total_returns.sort_values(ascending=False))
print(volatility.sort_values(ascending=False))
```
